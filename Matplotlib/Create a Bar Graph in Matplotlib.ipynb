```python
# Create a bar graph in Matplotlib
import matplotlib.pyplot as plt
time_sample = [5, 10, 15, 20, 25]
displacement_sample = [0.5, 1.0, 1.5, 1.0, 0.5]

plt.bar(time_sample, displacement_sample, color='orange', width=3)
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Displacement at Different Times During Earthquake')
plt.show()
```
