```python
import matplotlib.pyplot as plt
import numpy as np

#  random displacement values
data = np.random.normal(0, 0.2, 2000)

# Create the histogram
plt.hist(data, bins=40, color='pink', edgecolor='black')

mean_val = np.mean(data)
std_val = np.std(data)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.4f}m')
plt.axvline( std_val, color='blue', linestyle='solid', linewidth=2, label=f'Std Dev: {std_val:.4f}m')

plt.title('How much the building moved')
plt.xlabel('Displacement (meters)')
plt.ylabel('Ovserved Frequency (Hz)')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate  Earthquake Data
data = np.random.normal(loc=0, scale=0.2, size=50000)
df = pd.DataFrame({'Displacement': data})
print(df.head())
```

```python

mean_val = df['Displacement'].mean()
std_val = df['Displacement'].std()

plt.figure(figsize=(10, 6))
plt.hist(df['Displacement'], bins=40, color='violet', edgecolor='black')

plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')

plt.axvline(mean_val + std_val, color='blue', linestyle='solid', linewidth=2, label=f'SD (+1): {mean_val + std_val:.2f}')
plt.axvline(mean_val - std_val, color='blue', linestyle='solid', linewidth=2, label=f'SD (-1): {mean_val - std_val:.2f}')

plt.title('Displacement Histogram with Standard Deviation', fontsize=14, fontweight='bold')
plt.xlabel('Displacement (meters)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
```
