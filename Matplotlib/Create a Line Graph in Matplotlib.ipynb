```python
# Create a line graph in Matplotlib
import matplotlib.pyplot as plt
time_sample = [5, 10, 15, 20, 25]
displacement_sample = [0.5, 1.0, 1.5, 1.0, 0.5]
print("Time Sample:", time_sample)
print("Displacement Sample:", displacement_sample)

plt.plot(time_sample, displacement_sample, color='magenta')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Building Displacement Over Time')
plt.show()
```

```python
# Create a histogram in Matplotlib
plt.hist(displacement_sample, bins=20, color='orange', edgecolor='black')
plt.title('Frequency Distribution of Displacement During Earthquake')
plt.xlabel('Displacement (meters)')
plt.ylabel('Frequency')
plt.show()
```
