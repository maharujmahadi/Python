```python
# Create a Pie Chart in Matplotlib
import matplotlib.pyplot as plt
labels = ['Dhaka', 'Chattogram', 'Sylhet', 'Khulna', 'Rajshahi']
sizes = [40, 30, 25, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=90, colors=['red', 'green', 'blue', 'yellow', 'purple'])
plt.title('Seismic Event Phases')
plt.show()
```
