```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate earthquake data (displacement of a building)
time = np.linspace(0, 30, 300)  # 30 seconds with 300 points
displacement = np.sin(2 *np. pi * 0.1 * time) * np.exp(-0.05 * time)  # Sinusoidal decay function

# Convert to DataFrame
data = pd.DataFrame({'Time': time, 'Displacement': displacement})

# Show the first few rows of the data
print(data.head(15))
```

```python
# PyPlot Submodule - Plot a simple line graph
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Displacement'], label='Building Displacement')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Building Displacement During Earthquake')
plt.legend()
plt.show()
```

```python
# Plot a line with DataFrame
data.plot(x='Time', y='Displacement', label='Building Displacement', color='green', figsize=(10, 6))
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Building Displacement During Earthquake')
plt.legend()
plt.show()
```

```python
# Add grid lines to a plot
plt.figure(figsize=(9, 6))
plt.plot(data['Time'], data['Displacement'], label='Building Displacement', color='red')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Building Displacement During Earthquake')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Add labels to a plot
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Displacement'], label='Building Displacement', color='green')
plt.xlabel('Time (seconds)', fontsize=20)
plt.ylabel('Displacement (meters)', fontsize=12)
plt.title('Building Displacement During Earthquake', fontsize=14)
plt.legend()
plt.show()
```

```python
# Plot titles and position them
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Displacement'], label='Building Displacement', color='purple')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Building Displacement During Earthquake', fontsize=14, loc='left')  # Title positioned to the left
plt.legend()
plt.show()
```

```python
# Add a Matplotlib legend in a graph
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Displacement'], label='Building Displacement', color='orange')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Building Displacement During Earthquake')
plt.legend(title="Displacement Graph")
plt.show()
```

```python
# Change the background color of the legend
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Displacement'], label='Building Displacement', color='brown')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.title('Building Displacement During Earthquake')
plt.legend(facecolor='Red', edgecolor='Green')  # Custom background color
plt.show()
```
