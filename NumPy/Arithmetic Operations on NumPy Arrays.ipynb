```python
#Arithmetic Operations on NumPy Arrays
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

sum_arr = arr1 + arr2
print("Sum of arrays:", sum_arr)  # Output: [5 7 9] 
 
diff_arr = arr2 - arr1 
print("Difference of arrays:", diff_arr)  # Output: [3 3 3]

mul_arr = arr1 * arr2
print("Product of arrays:", mul_arr)  # Output: [ 4 10 18]
div_arr = arr2 / arr1
print("Division of arrays:", div_arr)  # Output: [4.  2.5 2.]
```
