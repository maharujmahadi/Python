```python
#Difference Between NumPy Arrays
import numpy as np
arr1 = np.array([1, 2, 3, 4])
print("array 1:", arr1)
arr2 = np.array([3, 4, 5, 6])
print("array 2:", arr2)

difference = np.setdiff1d(arr1, arr2)
print("difference:", difference)  # Output: [1 2]
```
