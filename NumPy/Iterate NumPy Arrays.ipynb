```python
import numpy as np
arr_2d = np.array([[1, 2], [3, 4], [5, 6]])
for row in arr_2d:
   print(row)  # Iterating through each row


```

```python
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
for block in arr_3d:
   for row in block:     
       print(row)  # Iterating through each block and row in 3D array
```

```python
arr_2d = np.array([[1, 2], [3, 4], [5, 6]])
for row in arr_2d:
    print(row)  # Iterating through each row

arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
for block in arr_3d:
    for row in block:
        print(row)  # Iterating through each block and row in 3D array
```
