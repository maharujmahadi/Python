```python
import numpy as np

# 2D Array
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
# Accessing specific element in 2D (Second row, third column)
print(arr_2d[1, 2])  # Output: 6

# 3D Array
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
# Accessing specific element in 3D (Second block, first row, second column)
print(arr_3d[1, 0, 1])  # Output: 6
```
