```python
#NumPy Array 
import numpy as np
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
# Accessing specific element in 2D
print(arr_2d[1, 2])  # Output: 6 (Second row, third column)

arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
# Accessing specific element in 3D
print(arr_3d[1, 0, 1])  # Output: 6 (Second block, first row, second column)
```
