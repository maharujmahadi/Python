```python
#NumPy Logs
import numpy as np
arr = np.array([1, 10, 100])
print("Original array:", arr)
log_arr = np.log10(arr)
print("Log10 of array:", log_arr)  
```

```python
#NumPy LCM and HCF
import numpy as np
a = np.array([12, 15])
print("Array a:", a)
b = np.array([18, 20])
print("Array b:", b)

lcm_arr = np.lcm(a, b)
gcd_arr = np.gcd(a, b)

print("LCM:", lcm_arr)  # Output: [36 60]
print("GCD:", gcd_arr)  # Output: [6 4]
```
