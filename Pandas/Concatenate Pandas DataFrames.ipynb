```python
# Concatenate DataFrames along rows
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print("DataFrame created:\n",df)
df3 = pd.DataFrame({'Name': ['Frank', 'Grace'], 'Age': [50, 55]})
concatenated_df = pd.concat([df, df3], ignore_index=True)

# Display the result
print("Concatenated DataFrame:\n", concatenated_df)
```
