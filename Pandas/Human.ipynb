```python
import turtle
import math

# 1. Setup
screen = turtle.Screen()
screen.setup(width=800, height=800)
screen.bgcolor("black")
t = turtle.Turtle()
t.speed(0)
turtle.tracer(0) 
t.hideturtle()

def draw_person(arm_angle, body_color):
    t.clear()
    t.pensize(5)
    t.pencolor(body_color)
    
    # Head
    t.penup()
    t.goto(0, 100)
    t.pendown()
    t.circle(40)
    
    # Body
    t.penup()
    t.goto(0, 100)
    t.pendown()
    t.goto(0, -50)
    
    # Arms (Moving)
    # Left Arm
    t.penup()
    t.goto(0, 70)
    t.pendown()
    t.goto(-60, 70 + arm_angle)
    
    # Right Arm
    t.penup()
    t.goto(0, 70)
    t.pendown()
    t.goto(60, 70 - arm_angle)
    
    # Legs
    t.penup()
    t.goto(0, -50)
    t.pendown()
    t.goto(-40, -150) # Left Leg
    t.penup()
    t.goto(0, -50)
    t.pendown()
    t.goto(40, -150) # Right Leg

# 2. Infinite Animation Loop
angle = 0
step = 0.05
colors = ["#FF5733", "#33FF57", "#3357FF", "#F333FF", "#FFFF33"]
color_index = 0

while True:
    # Calculate a waving motion using math.sin
    wave = math.sin(angle) * 40
    
    # Change color every few frames
    current_color = colors[int(color_index) % len(colors)]
    
    draw_person(wave, current_color)
    
    turtle.update()
    angle += step
    color_index += 0.01
```

```python
import turtle
import math

# 1. Setup
screen = turtle.Screen()
screen.setup(width=800, height=800)
screen.bgcolor("#AA1313") # Dark background for a cinematic feel
t = turtle.Turtle()
t.speed(0)
turtle.tracer(0)
t.hideturtle()

def draw_srk(pose_width):
    t.clear()
    
    # --- HAIR (Iconic SRK messy/thick hair) ---
    t.penup()
    t.goto(-50, 150)
    t.pendown()
    t.fillcolor("#000000")
    t.begin_fill()
    for _ in range(10):
        t.forward(100)
        t.left(160)
    t.end_fill()

    # --- FACE ---
    t.penup()
    t.goto(0, 50)
    t.pendown()
    t.pencolor("#ffdbac")
    t.pensize(3)
    t.circle(60)

    # --- SIGNATURE OPEN ARMS POSE ---
    t.pencolor("white")
    t.pensize(8)
    # Torso
    t.penup()
    t.goto(0, 50)
    t.pendown()
    t.goto(0, -100)
    
    # Left Arm (Animating Outwards)
    t.penup()
    t.goto(0, 30)
    t.pendown()
    t.goto(-20 - pose_width, 60 + (pose_width * 0.2))
    
    # Right Arm (Animating Outwards)
    t.penup()
    t.goto(0, 30)
    t.pendown()
    t.goto(20 + pose_width, 60 + (pose_width * 0.2))

    # --- LEGS ---
    t.goto(40, -200)
    t.penup()
    t.goto(0, -100)
    t.pendown()
    t.goto(-40, -200)

    # --- TEXT ---
    t.penup()
    t.goto(0, -280)
    t.color("gold")
    t.write("KING KHAN", align="center", font=("Arial", 30, "bold italic"))

# 2. Infinite Pose Loop
angle = 0
while True:
    # This math creates the smooth opening and closing of the arms
    spread = (math.sin(angle) + 1) * 60 
    
    draw_srk(spread)
    
    turtle.update()
    angle += 0.05
```
