```python
import math
from turtle import *

# 1. Setup
setup(width=800, height=800)
bgcolor("black")
pensize(4)
hideturtle()
tracer(0) 
colormode(255) # Allows us to use RGB colors for the gradient

def hearta(k, scale):
    return scale * 15 * math.sin(k) ** 3

def heartb(k, scale):
    return scale * (12 * math.cos(k) - 5 * math.cos(2 * k) - 2 * math.cos(3 * k) - math.cos(4 * k))

# 2. Growth Animation
scale = 0
growth_speed = 0.5  # Adjust this to make it spread faster or slower

while scale < 25:
    clear()
    
    # Dynamic color: gets brighter as it spreads
    red_val = int(min(255, 100 + (scale * 6))) 
    pencolor(red_val, 0, 50)
    
    penup()
    
    # Drawing the heart outline for the current scale
    t = 0
    while t < 2 * math.pi:
        x = hearta(t, scale)
        y = heartb(t, scale)
        goto(x, y)
        pendown()
        t += 0.05
    
    update()
    scale += growth_speed # This makes it "spread" outside

# Keep the final large heart on screen
done()
```

```python
import math
from turtle import *

# 1. Canvas Setup
setup (width=800, height=800)
bgcolor("black")
speed(0)
tracer(0) 
hideturtle()
colormode(255)

def hearta(k, scale):
    return scale * 15 * math.sin(k) ** 3

def heartb(k, scale):
    return scale * (12 * math.cos(k) - 5 * math.cos(2 * k) - 2 * math.cos(3 * k) - math.cos(4 * k))

# 2. Design Parameters
num_hearts = 36  # How many hearts to draw in the circle
scale = 15       # Size of each heart

for i in range(num_hearts):
    # Change color slightly for each heart to create a gradient
    color_val = int(255 * (i / num_hearts))
    pencolor(255, 40, color_val)
    pensize(2)
    
    penup()
    goto(0, 0)
    # Rotate the heading so they spread out in a circle
    setheading(i * (360 / num_hearts)) 
    
    # Draw one heart shape
    t = 0
    while t < 2 * math.pi:
        # We use rotation math to tilt the heart shapes
        angle = math.radians(i * (360 / num_hearts))
        
        # Original heart coordinates
        hx = hearta(t, scale)
        hy = heartb(t, scale)
        
        # Rotate coordinates around the center
        rx = hx * math.cos(angle) - hy * math.sin(angle)
        ry = hx * math.sin(angle) + hy * math.cos(angle)
        
        goto(rx, ry)
        pendown()
        t += 0.1
    
    update() # Show each heart as it is finished

done()
```
