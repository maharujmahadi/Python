```python
from scipy.fft import fft, ifft #fft (Fast Fourier Transform) and ifft (Inverse Fast Fourier Transform)
import numpy as np
import matplotlib.pyplot as plt

# Sample signal (a sine wave)
t = np.linspace(0, 1, 400)
signal = np.sin(2 * np.pi * 50 * t)

# Apply FFT
signal_fft = fft(signal)

# Apply IFFT to recover the original signal
signal_reconstructed = ifft(signal_fft)

# Plot the original and reconstructed signals
plt.subplot(2, 1, 1)
plt.plot(t, signal, label='Original Signal', color='orange',)
plt.legend(loc='upper right')
plt.subplot(2, 1, 2)
plt.plot(t, signal_reconstructed, label='Reconstructed Signal', linestyle='dashed', color='green')
plt.legend(loc='upper right')
plt.show()
```
