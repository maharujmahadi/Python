```python
import scipy.integrate
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

def model(t, y):
 return -2 * y + 1

t_span = (0, 5)
y0 = [0]
sol = solve_ivp(model, t_span, y0, dense_output=True)

t_vals = np.linspace(0, 5, 100)
y_vals = sol.sol(t_vals)
plt.plot(t_vals, y_vals[0], label='Solution of ODE')
plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
